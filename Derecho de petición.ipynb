```python
# INFORMACION PRELIMINAR

# En este campo se establece la información que inicialmente tiene que brindar el interesado.
from datetime import date
import unicodedata
pet_nom = input ('Ingrese sus nombres y apellidos\n').upper()
ciudad = input('¿En qué ciudad te encuentras? (Incluye el Departamento)\n').capitalize() 
fecha = date.today()
while True:
    ced_pet = input('¿Cuál es tu número de documento?:\n')
    ced_pet = ced_pet.replace ('.','').replace(",","")
    try:
        ced_pet = int (ced_pet)
    except:
        print('Solo puede introducir caracteres numericos. Ej.123456890. Por favor corrija los datos')
        continue
    ced_pet = str(ced_pet)
    if 6 < len (ced_pet) <= 10:
        break
    else:
        print('Verifique el valor ingresado. Recuerde que la cédula se compone por mínimo 6 digitos y máximo 10. Ej. 1234567890')
        continue
exped_pet = input('¿En qué ciudad se expidió tu documento de identidad?:\n').capitalize()
direccion = input('¿Cuál es tu dirección física?:\n')
#:::::::::::::::::::::::::::::::::::::::::::::::FUNCIONES:::::::::::::::::::::::::::::::::::::::::::::::::

def email_pedido(email_persona):
    while True:
        email = input(f'''¿Cuál es el correo electrónico del {email_persona}? \n''').lower()
        if '@' in email:
            break
        if '@' not in email:
            print('Ingrese un correo electrónico valido (Ej. @gmail.com)')
            continue
    return email

email_pet = email_pedido('propio')#poner como mejor va dirigida a
ado_nom = input('¿A quién va dirigido el Derecho de petición?: \n').upper()
direccion_ado =input('¿Cuál es la dirección física de la persona a quien va dirigido el Derecho de petición?:\n')

email_ado = email_pedido('peticionado')


while True:
    asunto = input('En mínimo una y máximo diez palabras, ¿cuál es el asunto de tu Derecho de petición?:\n').upper()# se peude limitar con un len y un while
    asunto = str (asunto)
    if 1 < len(asunto) <= 20:
        break
    else: 
        print ('Limite el uso de palabras. Recuerde que son máximo 10 palabras')
        continue

def contador(clase_escrito):
    hechos = []
    primer_hecho = input(f"""¿cuál es tu primer {clase_escrito}?\n""")
    hechos.append(primer_hecho)
    contador = 1 
    while True :
        while True: 
            otrohecho = input(f"""¿tienes otro {clase_escrito} que añadir? (si/no)\n""").lower()
            otrohecho = ''.join(filter(str.isalnum, otrohecho))
            otrohecho = unicodedata.normalize ("NFKD", otrohecho).encode("ascii", "ignore").decode("ascii")
            if otrohecho == 'no' or otrohecho == 'si':
                break 
            print ('Responda si o no')
        if otrohecho == 'si':
            contador = contador + 1
            cual_hecho = input(f'''describe el {clase_escrito} numero {contador}\n''')
            hechos.append(cual_hecho) 
            continue
        if otrohecho == 'no':
            break   
    return  hechos

    
#::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::   
# Hechos.
hechos = contador('hecho')

# Peticiones.
peticiones = contador('petición')

tipodoc = 'c.c'
#peticionario_tipodedocumento = input('¿Cuál es tu tipo de documento? (Puede ser Cédula de ciudadanía, Cédula de extranjería, Pasaporte o Permiso Especial de Permanencia): ')
calidad = 'ciudadano' #también puede ser servidor público etc
#peticionario_calidad = input('¿En qué calidad actuas? (Puede ser como ciudadano, como interesado, o en representación de otra persona, caso en el cual, deberás específicar el nombre y la identificación de tu representado): ')



#:::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::Petición:::::::::::::::::::::::::::::::::::::::

peticion_par1 = f"""{ciudad} {fecha}

Señores. 
{ado_nom}
{direccion_ado}
{email_ado}

Referencia: Derecho de Petición de {pet_nom} a {ado_nom} para {asunto}

{pet_nom}, identificado(a) con {tipodoc} número {ced_pet} de {exped_pet}, en calidad de {calidad}; de conformidad con el artículo 23 de la Constitución Política de Colombia de 1991, y del título segundo de la parte primera del Código de Procedimiento Administrativo y de lo Contencioso Administrativo (Ley 1437 de 2011); interpongo el siguiente Derecho de Petición basado en los siguientes:


i). Hechos.
{hechos}\n


ii). Petición.

En mérito de lo expuesto, se solicita a {ado_nom} que:

Petición 
{peticiones}\n
"""
peticion_par2= f'''iii). Pruebas.
Dentro del presente trámite se solicita se tengan como pruebas:
{pruebas}\n '''
peticion_par3 = f"""iv). Anexos.
Se anexan dentro del presente Derecho de petición todos los archivos relacionados en el acápite de pruebas, y los siguientes documentos.
{anexos}\n"""
peticion_par4 = f"""v). Documentos especiales requeridos por ley.
{documentos}\n"""
peticion_par5=f"""vi). Notificaciones.

Se podrá notificar en cualquiera de las siguientes direcciones.

{direccion}

{email_pet}


vii).Firma.

El presente documento se suscribe de conformidad con el artículo 7 de la Ley 527 de 1999, y con la presunción contemplada en el artículo 244 de la Ley 1564 de 2012 (Código General del Proceso).



Sin otro particular,



{pet_nom}
{tipodoc}- número {ced_pet}"""
#¿programar texto pruebas?


    

   
#pruebas

while True: 
    pruebas_acep = input('¿tienes que allegar una prueba? (si/no)\n').lower()
    pruebas_acep = ''.join(filter(str.isalnum, pruebas_acep))
    pruebas_acep = unicodedata.normalize ("NFKD", pruebas_acep).encode("ascii", "ignore").decode("ascii")
    if pruebas_acep == 'no' or pruebas_acep == 'si':
        break 
    print ('Responda si o no')
#anexos
while True: 
    anexo_acep = input ('¿tienes que allegar un anexo? (si/no)\n').lower()
    anexo_acep = ''.join(filter(str.isalnum, anexo_acep))
    
    anexo_acep = unicodedata.normalize ("NFKD", anexo_acep).encode("ascii", "ignore").decode("ascii")
    if anexo_acep == 'no' or anexo_acep == 'si':
        break 
    print ('Responda si o no')
# Documentos exigidos por ley.
while True: 
    doc_ley = input('¿tienes que allegar por ley un documento? (si/no)\n').lower()
    doc_ley = ''.join(filter(str.isalnum, doc_ley))
    doc_ley = unicodedata.normalize ("NFKD", doc_ley).encode("ascii", "ignore").decode("ascii")
    if doc_ley == 'no' or doc_ley == 'si':
        break 
    print ('Responda si o no')
    
if pruebas_acep == 'si' and anexo_acep == 'si' and doc_ley == 'no':
    
    anexos = contador('anexo')
    pruebas = contador('prueba')
    print(peticion_par1+peticion_par2+peticion_par3+peticion_par5)
    
elif anexo_acep == 'si' and  doc_ley == 'no' and  pruebas_acep == 'no':
    anexos = contador('anexo')
    print(peticion_par1+peticion_par3+peticion_par5)
elif doc_ley == 'si'and anexo_acep == 'no' and  doc_ley == 'no':
    documentos = contador ('documento')
    print(peticion_par1+peticion_par4+peticion_par5)
elif anexo_acep == 'si'and doc_ley == 'si' and pruebas_acep =='si':
    pruebas = contador('prueba')
    anexos = contador('anexo')
    documentos = contador ('documento por ley')
    print(peticion_par1+peticion_par2+ peticion_par3+peticion_par4+peticion_par5)
else:
    print(peticion_par1+peticion_par5)
```
