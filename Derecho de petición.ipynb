# DERECHO DE PETICIÓN

```python
#:::::::::::::::::::::::::::::::::::::::::::::LIBRERÍAS:::::::::::::::::::::::::::::::::::::::::::::::::::
from datetime import date #librería para evitar el input del usuario en cuanto la fecha
import unicodedata
#:::::::::::::::::::::::::::::::::::::::::::::::FUNCIONES:::::::::::::::::::::::::::::::::::::::::::::::::
def si_o_no(remplazo):# para que las pregunats asertivas las conteste
    while True:  # La pregunta ver como cuadre para las líneas 31
        doc= input(f'''¿tienes que allegar {remplazo}?(si/no)\n''').lower()
        doc = ''.join(filter(str.isalnum, doc))
        doc= unicodedata.normalize ("NFKD", doc).encode("ascii", "ignore").decode("ascii")
        if doc == 'no' or doc == 'si':
            break 
        print ('Responda si o no')
    return doc
    
def email_pedido(email_persona):
    while True:
        email = input(f'''¿Cuál es el correo electrónico del {email_persona}? \n''').lower()
        if '@' in email:
            break
        if '@' not in email:
            print('Ingrese un correo electrónico valido (Ej. @gmail.com)')
            continue
    return email

def contador(clase_escrito): #este sirve para añadir nuevos hechos; pruebas;peticiones etc. 
    hechos = [] #creamos una lsita a la cual se le añadirán los hechos
    primer_hecho = input(f"""¿cuál es tu primer {clase_escrito}?\n""")
    hechos.append(primer_hecho)
    contador = 1 
    while True :
        otrohecho = si_o_no(clase_escrito)
        if otrohecho == 'si':
            contador = contador + 1
            cual_hecho = input(f'''describe el {clase_escrito} numero {contador}\n''')
            hechos.append(cual_hecho) 
            continue
        if otrohecho == 'no':
            break   
    return  hechos
enum = 0
def enumerador() :
    global enum
    enum +=  1
    return enum
def organizador(lista): #para enumerar los hecjos; peticiones etc
        lista_ord = ''
        for i in  range(len(lista)):
            numerador = f'{i+1}. '+ lista[i] + '\n'
            lista_ord = lista_ord + numerador
        return lista_ord
#:::::::::::::::::::::::::::::::::::DICCIONARIO NUMERALES- LITERALES ROMANOS::::::::::::::::::::::::::::::::::::
literales = {1 :'I) ', 2: 'II) ', 3:'III) ',4:'V) ',5:'VI) ',6:'VI) ',7: 'VII) '}
# Para enumerar las partes del derecho de petición 

#::::::::::::::::::::::::::::::::::: CUERPO CÓDIGO::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::   
# INFORMACION PRELIMINAR
#-------------------------------------INFO PETICIONARIO------------------------------------------------

# En este campo se establece la información que inicialmente tiene que brindar el interesado.

pet_nom = input ('Ingrese sus nombres y apellidos\n').upper()
ciudad = input('¿En qué ciudad te encuentras? (Incluye el Departamento)\n').capitalize() 
fecha = date.today()#para poner la fecha del día en que se haga el poder
while True:
    ced_pet = input('¿Cuál es tu número de documento?:\n')
    ced_pet = ced_pet.replace ('.','').replace(",","")
    try:
        ced_pet = int (ced_pet)
    except:
        print('Solo puede introducir caracteres numericos. Ej.123456890. Por favor corrija los datos')
        continue
    ced_pet = str(ced_pet)
    if 6 < len (ced_pet) <= 10:
        break
    else:
        print('Verifique el valor ingresado. Recuerde que la cédula se compone por mínimo 6 digitos y máximo 10. Ej. 1234567890')
        continue
exped_pet = input('¿En qué ciudad se expidió tu documento de identidad?:\n').capitalize()
direccion = input('¿Cuál es tu dirección física?:\n')
email_pet = email_pedido('propio')#poner como mejor va dirigida a
tipodoc = 'c.c'
#peticionario_tipodedocumento = input('¿Cuál es tu tipo de documento? (Puede ser Cédula de ciudadanía, Cédula de extranjería, Pasaporte o Permiso Especial de Permanencia): ')
calidad = 'ciudadano' #también puede ser servidor público etc
#peticionario_calidad = input('¿En qué calidad actuas? (Puede ser como ciudadano, como interesado, o en representación de otra persona, caso en el cual, deberás específicar el nombre y la identificación de tu representado): ')


#--------------------------------------INFO PETICIONADO----------------------------------------------
ado_nom = input('¿A quién va dirigido el Derecho de petición?: \n').upper()
direccion_ado =input('¿Cuál es la dirección física de la persona a quien va dirigido el Derecho de petición?:\n')
email_ado = email_pedido('peticionado')

#-----------------------------------------ASUNTO-------------------------------------------
while True:
    asunto = input('En mínimo una y máximo diez palabras, ¿cuál es el asunto de tu Derecho de petición?:\n').upper()# se peude limitar con un len y un while
    asunto = str (asunto)
    if 1 < len(asunto) <= 20:
        break
    else: 
        print ('Limite el uso de palabras. Recuerde que son máximo 10 palabras')
        continue
#-----------------------------ELEMENTOS QUE VAN ENUMERADOS EN EL DERECHO--------------------------------
# Hechos.
hechos_lista = contador('hecho') 
hechos = organizador(hechos_lista) #los hechos organizados son los que quedan en el cuerpo del derecho 
enum_hechos = enumerador() # Para enumerar el acapite del derechod e petición
enum_hechos = literales[enum_hechos]# Para que sólo se asigne la clave correspondiente al número
# Peticiones.
peticiones_lista = contador('petición')
peticiones = organizador(peticiones_lista)
enum_pet = enumerador()
enum_pet = literales[enum_pet]
##||||||||||||||||||||||||||||||||||||||TEXTOS COMPLEMENTARIOS DEL DERECHO|||||||||||||||||||||||||||||||||||||||||
#pruebas
pruebas_txt = ''# para que no salga en el cuerpo si no se necesita
pruebas_acep = si_o_no('pruebas')
if pruebas_acep == 'si':
    pruebas_lista = contador('pruebas')
    pruebas = organizador(pruebas_lista)
    enum_pruebas = enumerador()
    enum_pruebas = literales[enum_pruebas]
    pruebas_txt =f'''{enum_pruebas}Pruebas. 
    Dentro del presente trámite se solicita se tengan como pruebas:
    {pruebas}\n '''
#anexos
anexos_txt = ''
anexos_acep= si_o_no('anexos')
if anexos_acep == 'si':
    anexos_lista = contador('anexos')
    anexos = organizador(anexos_lista)
    enum_anexos = enumerador()
    enum_anexos = literales[enum_anexos]
    anexos_txt =f"""\n{enum_anexos}Anexos.
    Se anexan dentro del presente Derecho de petición todos los archivos relacionados en el acápite de pruebas, y los siguientes documentos.
    {anexos}\n"""
#Documentos requeridos por ley
doc_txt = ''
doc_ley_acep = si_o_no('documento ley')
if doc_ley_acep == 'si':
    doc_ley_lista = contador('documento por ley')
    doc_ley = organizador(doc_ley_lista)
    enum_doc_ley = enumerador()
    enum_doc_ley = literales[enum_doc_ley]
    doc_txt = f"""\n{enum_doc_ley}Documentos especiales requeridos por ley.
    {doc_ley}\n"""
enum_notificaciones = enumerador()
enum_notificaciones = literales[enum_notificaciones]
enum_firma = enumerador()
enum_firma = literales[enum_firma]
#:::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::::Petición:::::::::::::::::::::::::::::::::::::::

peticion = f"""{ciudad} {fecha}

Señores. 
{ado_nom}
{direccion_ado}
{email_ado}

Referencia: Derecho de Petición de {pet_nom} a {ado_nom} para {asunto}

{pet_nom}, identificado(a) con {tipodoc} número {ced_pet} de {exped_pet}, en calidad de {calidad}; de conformidad con el artículo 23 de la Constitución Política de Colombia de 1991, y del título segundo de la parte primera del Código de Procedimiento Administrativo y de lo Contencioso Administrativo (Ley 1437 de 2011); interpongo el siguiente Derecho de Petición basado en los siguientes:


{enum_hechos}Hechos.
{hechos}\n


{enum_pet}Petición.

En mérito de lo expuesto, se solicita a {ado_nom} que:

Petición 
{peticiones}\n
{pruebas_txt}
{anexos_txt}
{doc_txt}
\n{enum_notificaciones}Notificaciones.

Se podrá notificar en cualquiera de las siguientes direcciones.

{direccion}

{email_pet}


{enum_firma}Firma.

El presente documento se suscribe de conformidad con el artículo 7 de la Ley 527 de 1999, y con la presunción contemplada en el artículo 244 de la Ley 1564 de 2012 (Código General del Proceso).



Sin otro particular,



{pet_nom}
{tipodoc} número {ced_pet}"""
print(peticion)
```
